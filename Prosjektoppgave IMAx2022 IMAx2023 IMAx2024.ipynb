## Del 1: Trafikkmodell: tetthet, hastighet og fluks

Vi skal modellere trafikkflyten langs en veistrekning. Vi innfører flere forenklinger:

- det er ingen veikryss 
- ingen forbikjøringer
- kun én kjørefil
- ingen kjører ut av eller inn på veien
- alle bilene er like lange
- trafikken flyter én vei
- alle sjåførene har samme kjøreadferd, som inkluderer at alle holder seg til fartsgrensen.
- m. m.

Merk at det er mulig å modellere trafikken med et system av ordinære differensiallikninger hvor vi regner på posisjon, hastighet og akselerasjon til hver enkelt bil. Det ville vært en såkalt *mikroskopisk* modell.

Det er derimot *ikke* hva vi skal gjøre i denne oppgaven. Vi skal se på en *makroskopisk* modell.

Vi modellerer trafikken som den ser ut observert fra siden av veien. Det observeres altfor mange biler til at vi kan se på hver enkelt bil. Så vi beskriver trafikken ved hjelp av funksjonen $u(x,t)$, som gir tettheten av biler i posisjon $x$ langs veien ved tidspunkt $t$. Posisjonene til startpunkt og endepunkt av veien vil vi kalle for hhv. $a$ og $b$, slik at $x\in[a,b]$.

Vi tenker oss at $u(x,t)$ for et tidspunkt $t$ er gitt ved antall biler på strekning $[x,x+R]$ delt på lengden av strekningen $R$, hvor $R$ er en passende lengde.

Måleenhet for $u(x,t)$ er derfor antall biler per meter. Vi har alltid $u(x,t)\geq 0.$

Hastigheten $v(x,t)$ til bilene ved posisjon $x$ og tidspunkt $t$ er også viktig.

#### Vår viktigste antakelse er at hastigheten $v(x,t)$ kun er avhengig av tettheten $u(x,t)$. Dvs.  hastigheten $v=v(u)$ er en funksjon av $u$.

For hastigheten $v$ så antar vi i tillegg at:

1. $v\geq 0$  
2. $v\leq v_{max}$, hvor $v_{max}$ er maksfart, dvs. fartsgrensen på veien.
3. $\frac{dv}{du}\leq 0$.
4. $v(0) = v_{max}$ når tettheten $u=0$.
5. $v(u_{max}) = 0$ når tettheten $u=u_{max}$ er størst mulig. 

Det er mange mulige valg av funksjoner $v=v(u)$. I praksis er dette noe vi må måle på den aktuelle veistrekningen. Det blir derimot for komplisert og tidkrevende for oss. Istedet prøver vi med noen funksjoner som kan passe.

Den enklest mulige funksjonen er den lineære funksjonen 
$$
\begin{align}\tag{1}
  v=v_{max}\left(1-\frac{u}{u_{max}}\right).
  \end{align}
$$

En annen mulig funksjon er den kvadratiske funksjonen
$$
\begin{align}\tag{2}
  v=v_{max}\left(1-\left(\frac{u}{u_{max}}\right)^2\right).
  \end{align}
$$

Dette er på ingen måte de eneste mulighetene for $v(u)$. Vi gir flere eksempler lenger nede i oppgaven, og vi vil også at dere skal prøve å lage noen slike funksjoner.

I oppgaven kommer vi hovedsaklig til å anta at funksjonen $v(u)$ er kjent. Hvis vi hadde brukt modellen i virkeligheten så måtte vi ha gått ut og målt hva $v$ er som funksjon av tettheten $u$. Og da ville lengden $R$ ha spilt en viktig rolle. Men siden vi bare antar hva $v(u)$ er, så spiller ikke lengden $R$ noen stor rolle.

Det samme gjelder for bestemmelsen av initial- og randbetingelsene i modellen: Hvis vi hadde brukt modellen i virkeligheten så måtte vi ha målt initial- og randbetingelsene, og også der ville lengden $R$ ha spilt en viktig rolle. Men vi kommer til å anta hva initial- og randbetingelsene er, og derfor spiller ikke lengden $R$ noen stor rolle der heller.

Vi skal også se på strømmen eller *fluksen* av biler gjennom et punkt $x$ langs veien ved tiden $t$. 

Vi definerer fluksen av biler, $J(x,t)$, til å være antall biler som passerer et punkt $x$ per sekund. Altså har $J$ måleenhet antall biler per sekund. I vår modell antar vi følgende sammenheng mellom tettheten $u$, hastigheten $v$ og fluksen $J$:
$$
\begin{align}\tag{3}
  J(x,t)=J(u(x,t))=u(x,t)v(u(x,t)).
\end{align}
$$
og vi krever at $J(x,t)\geq 0$.

Antall biler som passerer punktet $x$ på veien i tidsintervallet $[t,t+\Delta t]$ er altså (tilnærmet lik)
$$
\begin{align}\tag{4}
  J(x,t)\Delta t
  \end{align}
$$

Videre er antall biler som befinner seg på strekningen $[x,x+\Delta x]$ ved tiden $t$ gitt ved (tilnærmet) 
$$
\begin{align}\tag{5}
  u(x,t)\Delta x
\end{align}
$$

Nå skal vi sette opp et regnskap for bilene, som er helt analogt med det de fleste av oss har et forhold til. Vi vet at penger inn på konto minus penger ut av konto over et tidsrom er lik netto endring av saldoen.

Tilsvarende er antall biler på strekningen $[x,x+\Delta x]$ ved tiden $t+\Delta t$ minus antall biler på strekningen $[x,x+\Delta x]$ ved tiden $t$ lik netto endring i antall biler på strekningen $[x,x+\Delta x]$ i tidsrommet $t$ til $t+\Delta t$.

På samme måte som med bankkontoen, må dette være lik antall biler som kjører inn på strekningen $[x,x+\Delta x]$ minus antall biler som kjører ut fra strekningen $[x,x+\Delta x]$ i tidsrommet fra $t$ til $t+\Delta t$.

Uttrykker vi regnskapet matematisk, så fører det til ligningen
$$
\begin{align}\tag{6}
    u(x,t+\Delta t)\cdot \Delta x-u(x,t)\cdot \Delta x=-J(x+\Delta x, t)\cdot \Delta t + J(x,t)\cdot \Delta t.
\end{align}
$$
som etter deling på $\Delta x$ og $\Delta t$ gir
$$
\begin{align}\tag{7}
    \frac{u(x,t+\Delta t)-u(x,t)}{\Delta t}=-\frac{J(x+\Delta x, t)-J(x,t)}{\Delta x}.
    \end{align}
$$
Som vanlig i denne typen modellering ser vi på ligningen over når $\Delta x\rightarrow 0$ og $\Delta t\rightarrow 0$. 

Ligningen over blir da den partielle differensiallikningen
$$
\begin{align}\tag{8}
    \frac{\partial u(x,t)}{\partial t}=-\frac{\partial J(x,t)}{\partial x}
    \end{align}
$$
som gjelder for verdier $x\in [a,b]$ og $t\geq 0$.  

Som vi har sett ovenfor, er $J$ er en funksjon av $u$ gitt ved $J(u)=u\cdot v(u)$, slik at likningen til slutt kan skrives som 
$$
\begin{align}\tag{9}
  u_t + (u\cdot v(u))_x = 0
\end{align}
$$

I tillegg så vil en fullstendig problembeskrivelse inneholde mer informasjon. For eksempel så beskriver randbetingelsen $u(a,t)=u_a(t)$ hvor stor tettheten av biler er til enhver tid ved startpunktet av veien $(a)$, mens initialbetingelsen $u(x,0)=u_0(x)$ beskriver tettheten av biler på alle posisjoner ved starttid $t=0$.

Nå skal vi se på numerisk løsning av Poissonligningen i 2 dimensjoner

$$
\begin{align}\tag{17}
u_{xx} + u_{yy} = f(x,y), \quad -5<x<5, 0<y<2.
\end{align}
$$

Vi velger $f=0$ og følgende randbetingelser:

$$
\begin{align}\tag{18}
u(-5,y) = \sin(2\pi y), \quad & 0\leq y \leq 2 \\
u(x,0) = 0, \quad & -5\leq x \leq 5 \\
u(5,y) = \sin(2\pi y), \quad & 0\leq y \leq 2 \\
u(x,2) = \sin(\pi x), \quad & -5 \leq x \leq 5
\end{align}
$$

Løs ligningen numerisk.

## Oppgave 7: Luften skal med (10 poeng)

Vår antagelse at temperaturen i luften holdes konstant ved $200$ grader er ikke realistisk. I realiteten vil temperaturen i ovnen gå ned når et kaldere legeme settes inn i ovnen. 

Det vil også oppstå konveksjon, men dette er noe vi vil ignorere i denne oppgaven.

Vi tar et lag av luft rundt legemet med i modellen. I luftlaget så har vi fortsatt den partielle differensiallikningen
$$
\begin{align}\tag{20}
u_t = \alpha \cdot ( u_{xx} + u_{yy} ),
\end{align}
$$
hvor $\alpha$ tar en annen lavere verdi som passer bedre til luft.

På randen rundt luftlaget har vi fortsatt konstant temperatur $200$ for alle verdier av $t$.

I luftlaget så er temperaturen $200$ grader når $t=0$. Temperaturen til legemet vi setter inn ved $t=0$ er $15$ grader.

Bredden og høyden av luftlaget skal være omtrent halvparten av bredden og høyden til legemet. Dere kan selv bestemme de nøyaktige dimensjonene.
